
# Plot Transformation

We can display transformations by plotting the basis vectors of the
corresponding coordinate frame.


```python
import matplotlib.pyplot as plt

from pytransform3d.plot_utils import make_3d_axis
from pytransform3d.transformations import plot_transform

ax = make_3d_axis(ax_s=1, unit="m", n_ticks=6)  # ax_s:缩放新的 matplotlib 坐标轴 unit:单位；n_ticks:每个轴上的刻度数
plot_transform(ax=ax)  # pr.plot_tramsform:绘制变换；ax:matplotlib 3d 轴
plt.tight_layout()
plt.show()
```
